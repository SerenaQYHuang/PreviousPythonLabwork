```python
import jieba
import pandas as pd
```

```python
fh=pd.read_csv('news200.txt','r',delimiter='\t',header=0,encoding='utf-8')
```

```python
fh.head()
```

```python
fh.tail()
```

```python
test='初中学历身体健康月薪6500 这家企业却招不到人'
list(jieba.cut(test))
```

```python
t_list=[]
for i in fh['title']:
    seg=list(jieba.cut(i))
    t_list.append(seg)
    
print(t_list)
```

```python
t_list2=[]
for i in t_list:
    combined_title=' '.join(i)
    t_list2.append(combined_title)
    
print(t_list2)
```

```python
df=pd.DataFrame(list(zip(fh['category'],t_list2)),columns=['category','title'])
df.head()
```

```python
# fist define x and y
x=df.title
y=df.category
print(x.shape)
```

```python
# split the data into training and testing set

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1)
```

```python
# instantiate the vectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer()
```

```python
vect.fit(xtrain)
xtrain_dtm=vect.transform(xtrain)
```

```python
xtrain_dtm
```

```python
# transform the testing into a document-term matrix
xtest_dtm=vect.transform(xtest)
xtest_dtm
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
```

```python
# train the model using xtrain data
%time nb.fit(xtrain_dtm,ytrain)
```

```python
# make class predictions for x test dtm

%time ypred_class=nb.predict(xtest_dtm)
```

```python
# calculate the accuracy of class prediction
from sklearn import metrics
metrics.accuracy_score(ytest,ypred_class)
```

```python
# confusion matrix
metrics.confusion_matrix(ytest,ypred_class)
```

```python
# find out the false positive (entertainment incorrectly classified as social news)

xtest[(ypred_class==1)&(ytest==0)]
```

```python
# find out the social news incorrectly predicted as entertainment
xtest[(ypred_class==0)&(ytest==1)]
```
