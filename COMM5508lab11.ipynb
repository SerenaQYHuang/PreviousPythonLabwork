```python
# 
from time import time
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.decomposition import NMF,LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
```

```python
n_samples=2000
n_features=1000
n_topics=10
n_top_words=20
```

```python
dataset=fetch_20newsgroups(shuffle=True,random_state=1,remove=('headers','footers','quotes'))
```

```python
dataset.data[0]
```

```python
vectorizer=TfidfVectorizer(max_df=0.95,min_df=2,max_features=n_features,stop_words='english',ngram_range=(1,1))
tfidf=vectorizer.fit_transform(dataset.data[:n_samples])
```

```python
nmf=NMF(n_components=n_topics,random_state=1).fit(tfidf)
```

```python
nmf.components_
```

```python
nmf.components_[0]
```

```python
topic1=sorted(nmf.components_[0],reverse=True)
print(topic1)
```

```python
feature_names=vectorizer.get_feature_names()
feature_names[list(nmf.components_[0]).index(topic1[0])]
```

```python
feature_names=vectorizer.get_feature_names()
topic=[]
topid=0
while topid<n_topics:
    topic=sorted(nmf.components_[topid],reverse=True)
    wordid=0
    wordlist=[]
    while wordid<n_top_words:
        wordlist.append(feature_names[list(nmf.components_[topid]).index(topic[wordid])])
        wordid+=1
    topid+=1
    print('feature word for tipic %d'%topid)
    print(wordlist)
```
